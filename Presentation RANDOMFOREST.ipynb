<h3 style="color:rgb(48,80,186);
 text-align:center;font-family: Open Sans, monospace;   font-size:20px  
" >
I-Définition de la Forêt Aléatoire</h3>

<p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold
">La Forêt Aléatoire est un algorithme d'apprentissage automatique faisant partie de la famille des méthodes d'ensemble. Elle est utilisée pour des tâches de classification, de régression et d'autres problèmes de prédiction. La Forêt Aléatoire construit plusieurs arbres de décision pendant l'entraînement et combine leurs prédictions pour obtenir une réponse plus robuste et précise.</p>

<h3 style="color:rgb(48,80,186);
 text-align:center;font-family: Open Sans, monospace;   font-size:20px  
" >
II-Principes de Base</h3>

<p  style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold"><strong style="color:rgb(12,33,187);font-family: 'Courier New', Courier, monospace;">- Arbres de Décision :</strong> La Random Forest repose sur l'utilisation d'arbres de décision comme composants de base. Un arbre de décision est une structure arborescente où chaque nœud représente une décision basée sur une caractéristique, et chaque feuille représente une prédiction.</p>
  <p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold;margin-top:20px"><strong style="color:rgb(12,33,187);font-family: 'Courier New', Courier, monospace;">- Ensemble d'Arbres :</strong> La Random Forest construit un ensemble d'arbres de décision. Chaque arbre est construit sur un sous-ensemble aléatoire des données d'entraînement et utilise un sous-ensemble aléatoire des caractéristiques (variables) à chaque nœud de décision.</p>

  <p  style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold;margin-top:20px"><strong style="color:rgb(12,33,187);font-family: 'Courier New', Courier, monospace;">- Vote Majoritaire :</strong> Pour la classification, la prédiction finale de la Random Forest est déterminée par un vote majoritaire des arbres. Chaque arbre "vote" pour une classe, et la classe avec le plus grand nombre de votes devient la prédiction finale.</p>

   <p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold;margin-top:20px"><strong style="color:rgb(12,33,187);font-family: 'Courier New', Courier, monospace;">- Moyenne ou Somme :</strong> Pour la régression, la prédiction finale est souvent obtenue en prenant la moyenne (pour les valeurs continues) ou la somme (pour les valeurs discrètes) des prédictions des arbres individuels.</p>

<h3 style="color:rgb(48,80,186);
 text-align:center;font-family: Open Sans, monospace;   font-size:20px  
">
III-Pourquoi la Forêt Aléatoire?</h3>

<ul>
    <li  style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold"><strong style="color:rgb(12,33,187);font-family: 'Courier New', Courier, monospace;">Importance des Caractéristiques :</strong> La Forêt Aléatoire peut traiter un grand nombre de variables et fournir des indications sur l'importance de chaque variable dans la réalisation des prédictions. Cela est crucial pour comprendre quelles variables contribuent le plus à la précision prédictive.</li>
   <li style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold;margin-top:20px"><strong style="color:rgb(12,33,187);font-family: 'Courier New', Courier, monospace;">Apprentissage en Ensemble :</strong> La Forêt Aléatoire est une méthode d'apprentissage en ensemble, combinant les prédictions de plusieurs arbres de décision. Cela aide à réduire le surajustement et à améliorer la robustesse globale du modèle.</li>
   <li style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold ;margin-top:20px" ><strong style="color:rgb(12,33,187);font-family: 'Courier New', Courier, monospace;">Robustesse face aux Données Bruitées:</strong>Robustesse face aux Données Bruitées :</strong> Les Forêts Aléatoires sont intrinsèquement robustes face aux données bruitées et aux valeurs aberrantes. La randomisation introduite lors de la construction des arbres individuels aide à atténuer l'impact des observations bruitées.</li>
   <li style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold;margin-top:20px"><strong style="color:rgb(12,33,187);font-family: 'Courier New', Courier, monospace;">Gestion des Valeurs Manquantes  :</strong> Gestion des Valeurs Manquantes :</strong> Les Forêts Aléatoires peuvent gérer les valeurs manquantes dans l'ensemble de données sans nécessiter d'imputation. Cela peut être bénéfique lors de la manipulation d'ensembles de données réels où les données manquantes sont courantes..</li>


   
</ul>

Le Bootstrap</h3>

<p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold
">Pour chaque arbre de l'ensemble, un sous-ensemble aléatoire de l'ensemble de données d'origine est créé grâce à un processus appelé "bootstrapping". Le bootstrapping consiste à sélectionner des échantillons aléatoires avec remplacement à partir de l'ensemble de données d'origine. Cela signifie que certains points de données apparaîtront plusieurs fois dans l'échantillon bootstrap, tandis que d'autres peuvent être omis complètement..</p>

```python
random.seed(0)
# Number of bootstrap samples to create <=>  number of decisionTrees
num_samples = 4

# Randomly select data points with replacement to create bootstrap samples
random_samples = [random.choices(data, k=6) for _ in range(num_samples)]

# Definig colors for rows (colors for data rows)
colors = [Fore.RED, Fore.GREEN, Fore.BLUE, Fore.YELLOW, Fore.MAGENTA, Fore.LIGHTCYAN_EX,Fore.LIGHTYELLOW_EX,Fore.LIGHTGREEN_EX,Fore.BLACK]

for i, sample in enumerate(random_samples):
    table = tabulate(pd.DataFrame(sample), headers='keys', tablefmt='pretty', showindex=False)
    for i, line in enumerate(table.split('\n')):
         if i > 0:  # Skip the header line
                colored_line = f"{colors[i - 1]}{line}{Style.RESET_ALL}"
                print(colored_line)
        
         else:
            print(line)

   
    print()

```

<h3 style="color:rgb(48,80,186);
 text-align:center;font-family: Open Sans, monospace;   font-size:25px  
">
Sélection Aléatoire des Caractéristiques  "feature subsampling"</h3>

<p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold
">Alors que les échantillons bootstrap sont différents pour chaque arbre, le processus de sélection des caractéristiques est également aléatoire. À chaque nœud de l'arbre de décision lors de sa construction, un sous-ensemble aléatoire de caractéristiques est pris en compte pour diviser le nœud. La taille de ce sous-ensemble aléatoire est généralement la racine carrée du nombre total de caractéristiques. Cette sélection aléatoire des caractéristiques introduit une diversité supplémentaire entre les arbres.</p>

```python

# storing data of samples
samples=[]

# Fixer la graine aléatoire pour la reproductibilité

random.seed(0)
# Nombre d'échantillons bootstrap à créer, c'est-à-dire le nombre d'arbres de décision
num_samples = 4

# Définir le nombre total de caractéristiques dans votre jeu de données (en supprimant la colonne "ID" et "Target (y)")
total_features = len(data[0]) - 2

# Boucle pour chaque échantillon bootstrapa
for _ in range(num_samples):
    # Calculer le nombre de caractéristiques à choisir pour cet échantillon (la racine carrée du nombre total de caractéristiques)
    num_features_to_select = int(math.sqrt(total_features))

    # Sélection aléatoire des indices de caractéristiques (en ajoutant 1 à total_features pour la cible)
    selected_feature_indices = random.sample(range(1, total_features + 1), num_features_to_select)

    # Ajouter l'indice correspondant à la colonne 'y' à la liste des indices sélectionnés
    selected_feature_indices.append(total_features + 1)  # Index for the 'y' target variable

    # Sélection aléatoire d'individus avec remplacement pour créer l'échantillon de la taille spécifiée (6 individus)
    selected_individuals = [random.choice(data) for _ in range(6)]

    # Créer un sous-ensemble de données avec les caractéristiques sélectionnées pour cet échantillon, y compris la cible 'y'
    selected_data = [{k: v for i, (k, v) in enumerate(sample.items()) if i == 0 or i in selected_feature_indices} for sample in selected_individuals]
    samples.append(selected_data)
    # Définir les couleurs pour les lignes (couleurs pour les lignes de données)
    colors = [Fore.RED, Fore.GREEN, Fore.BLUE, Fore.YELLOW, Fore.MAGENTA, Fore.LIGHTCYAN_EX, Fore.LIGHTYELLOW_EX, Fore.LIGHTGREEN_EX, Fore.BLACK]

    # Afficher les données de l'échantillon bootstrap
    table = tabulate(pd.DataFrame(selected_data), headers='keys', tablefmt='pretty', showindex=False)
    
    for i, line in enumerate(table.split('\n')):
         if i > 0:  # Skip the header line
                colored_line = f"{colors[i - 1]}{line}{Style.RESET_ALL}"
                print(colored_line)
        
        
    print()
   
```

<h3 style="color:rgb(48,80,186);
 text-align:center;font-family: Open Sans, monospace;   font-size:20px  
">
IV-Approche Mathematique</h3>

 <p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold">La Forêt Aléatoire construit un ensemble d'arbres de décision. Chaque arbre est construit sur un sous-ensemble aléatoire des données d'entraînement et utilise un sous-ensemble aléatoire des caractéristiques (variables) à chaque nœud de décision. Cela introduit de la diversité dans les arbres.</p>

<p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold">Soit $(T)$ le nombre total d'arbres dans la forêt, la prédiction de la forêt pour une observation $(X)$est donnée par la moyenne (pour la régression) ou le vote majoritaire (pour la classification) des prédictions des arbres individuels :</p>
 <p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold">
    $\hat{Y}(X) = \frac{1}{T} \sum_{i=1}^{T} f_i(X)$
  </p>

  <p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold">
    Cela représente la prédiction de la Forêt Aléatoire pour une observation \(X\). Expliquons brièvement chaque composant de l'expression :
  </p>

  <ul>
    <li style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold">$(\hat{Y}(X))$ : La prédiction de la Forêt Aléatoire pour l'observation \(X\).</li>
