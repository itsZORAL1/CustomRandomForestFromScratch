<p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold

II-Principes de Base</h3>

<p  style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold"><strong style="color:rgb(12,33,187);font-family: 'Courier New', Courier, monospace;">- Arbres de Dcision :</strong> La Random Forest repose sur l'utilisation d'arbres de dcision comme composants de base. Un arbre de dcision est une structure arborescente o chaque nud reprsente une dcision base sur une caractristique, et chaque feuille reprsente une prdiction.</p>
  <p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold;margin-top:20px"><strong style="color:rgb(12,33,187);font-family: 'Courier New', Courier, monospace;">- Ensemble d'Arbres :</strong> La Random Forest construit un ensemble d'arbres de dcision. Chaque arbre est construit sur un sous-ensemble alatoire des donnes d'entranement et utilise un sous-ensemble alatoire des caractristiques (variables)  chaque nud de dcision.</p>

  <p  style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold;margin-top:20px"><strong style="color:rgb(12,33,187);font-family: 'Courier New', Courier, monospace;">- Vote Majoritaire :</strong> Pour la classification, la prdiction finale de la Random Forest est dtermine par un vote majoritaire des arbres. Le Bootstrap</h3>

<p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold
">Pour chaque arbre de l'ensemble, un sous-ensemble alatoire de l'ensemble de donnes d'origine est cr grce  un processus appel "bootstrapping". Le bootstrapping consiste  slectionner des chantillons alatoires avec remplacement  partir de l'ensemble de donnes d'origine. Cela signifie que certains points de donnes apparatront plusieurs fois dans l'chantillon bootstrap, tandis que d'autres peuvent tre omis compltement..</p>

```python
random.seed(0)
# Number of bootstrap samples to create <=>  number of decisionTrees
num_samples = 4

# Randomly select data points with replacement to create bootstrap samples
random_samples = [random.choices(data, k=6) for _ in range(num_samples)]

# Definig colors for rows (colors for data rows)
colors = [Fore.RED, Fore.GREEN, Fore.BLUE, Fore.YELLOW, Fore.MAGENTA, Fore.LIGHTCYAN_EX,Fore.LIGHTYELLOW_EX,Fore.LIGHTGREEN_EX,Fore.BLACK]

for i, sample in enumerate(random_samples):
    table = tabulate(pd.DataFrame(sample), headers='keys', tablefmt='pretty', showindex=False)
    for i, line in enumerate(table.split('\n')):
         if i > 0:  # Skip the header line
                colored_line = f"{colors[i - 1]}{line}{Style.RESET_ALL}"
                print(colored_line)
        
         else:
            print(line)

   
    print()

```

Slection Alatoire des Caractristiques  "feature subsampling"</h3>

<p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold
">Alors que les chantillons bootstrap sont diffrents pour chaque arbre, le processus de slection des caractristiques est galement alatoire.  chaque nud de l'arbre de dcision lors de sa construction, un sous-ensemble alatoire de caractristiques est pris en compte pour diviser le nud. La taille de ce sous-ensemble alatoire est gnralement la racine carre du nombre total de caractristiques. Cette slection alatoire des caractristiques introduit une diversit supplmentaire entre les arbres.</p>

```python

# storing data of samples
samples=[]

# Fixer la graine alatoire pour la reproductibilit

random.seed(0)
# Nombre d'chantillons bootstrap  crer, c'est--dire le nombre d'arbres de dcision
num_samples = 4

# Dfinir le nombre total de caractristiques dans votre jeu de donnes (en supprimant la colonne "ID" et "Target (y)")
total_features = len(data[0]) - 2

# Boucle pour chaque chantillon bootstrapa
for _ in range(num_samples):
    # Calculer le nombre de caractristiques  choisir pour cet chantillon (la racine carre du nombre total de caractristiques)
    num_features_to_select = int(math.sqrt(total_features))

    # Slection alatoire des indices de caractristiques (en ajoutant 1  total_features pour la cible)
    selected_feature_indices = random.sample(range(1, total_features + 1), num_features_to_select)

    # Ajouter l'indice correspondant  la colonne 'y'  la liste des indices slectionns
    selected_feature_indices.append(total_features + 1)  # Index for the 'y' target variable

    # Slection alatoire d'individus avec remplacement pour crer l'chantillon de la taille spcifie (6 individus)
    selected_individuals = [random.choice(data) for _ in range(6)]

    # Crer un sous-ensemble de donnes avec les caractristiques slectionnes pour cet chantillon, y compris la cible 'y'
    selected_data = [{k: v for i, (k, v) in enumerate(sample.items()) if i == 0 or i in selected_feature_indices} for sample in selected_individuals]
    samples.append(selected_data)
    # Dfinir les couleurs pour les lignes (couleurs pour les lignes de donnes)
    colors = [Fore.RED, Fore.GREEN, Fore.BLUE, Fore.YELLOW, Fore.MAGENTA, Fore.LIGHTCYAN_EX, Fore.LIGHTYELLOW_EX, Fore.LIGHTGREEN_EX, Fore.BLACK]

    # Afficher les donnes de l'chantillon bootstrap
    table = tabulate(pd.DataFrame(selected_data), headers='keys', tablefmt='pretty', showindex=False)
    
    for i, line in enumerate(table.split('\n')):
         if i > 0:  # Skip the header line
                colored_line = f"{colors[i - 1]}{line}{Style.RESET_ALL}"
                print(colored_line)
        
        
    print()
   
```

<h3 style="color:rgb(48,80,186);

IV-Approche Mathematique</h3>

 <p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold">La Fort Alatoire construit un ensemble d'arbres de dcision. Chaque arbre est construit sur un sous-ensemble alatoire des donnes d'entranement et utilise un sous-ensemble alatoire des caractristiques (variables)  chaque nud de dcision. Cela introduit de la diversit dans les arbres.</p>

<p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold">Soit $(T)$ le nombre total d'arbres dans la fort, la prdiction de la fort pour une observation $(X)$est donne par la moyenne (pour la rgression) ou le vote majoritaire (pour la classification) des prdictions des arbres individuels :</p>
 <p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold">
    $\hat{Y}(X) = \frac{1}{T} \sum_{i=1}^{T} f_i(X)$
  </p>

  <p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold">
    Cela reprsente la prdiction de la Fort Alatoire pour une observation \(X\). Expliquons brivement chaque composant de l'expression :
  </p>

  <ul>
    <li style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold">$(\hat{Y}(X))$ : La prdiction de la Fort Alatoire pour l'observation \(X\).</li>
    <li style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold">$(T)$ : Le nombre total d'arbres dans la fort.</li>
    <li style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold">$(f_i(X))$ : La prdiction de l'arbre \(i\) pour l'observation \(X\).</li>
    <li style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold">$\sum_{i=1}^{T}$ : La sommation sur tous les arbres de la fort.</li>
  </ul>

 <p style="color: black; font-family:  Open sans, monospace;font-size:16px;font-weight:bold">
